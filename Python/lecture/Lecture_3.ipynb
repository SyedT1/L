## **Operators And Expression**
An _operator_ is a character or characters in python will 
determine how certain computations are carried out.  
_Operators_ act on values which are known as operands.   
##### For example:  
+ operand (_operator_) operand
+ (_operator_) operand  
```python
  a = 10
  b = 2
  result = a*b
  print("the result is {0}".format(result))
  ```
Here, a * b is an expression  where * is an _operator_ and a and b are operands.
```python
 a = -10
 print("the value of a is {0}".format(-a))
 ```

These are the examples of an **Expression** . An expression is a sequence of 
 operators, variables, constants and function calls that results in a value.  

##### There are **7** types of operators in python:
+ Arithmetic operators
+ Relational operators
+ Assignment operators
+ Logical operators
+ Membership operators
+ Identify operators
+ [Bitwise operators](https://en.wikipedia.org/wiki/Bitwise_operation)

Let's start with **Arithmetic operators**.   

### Unary operators
Those operators that work with just a single operand are called _unary operators_ which is (_operator_)operand  
   + +var :  
        The + operator can be utilized in unary form before the operand.But, it 
        doesn't change the operand.
        ```python 
           a = 10
           print("the value of a is {0} and it's type is {1}".format(a,type(a)))
        ```
   + -var :  
        The - operator can also be utilized in unary form before the operand. It
        basically negates the value of the operand. Change also depends on the 
        value of the operand.
        ```python
           b = 10
           print("the value of b is {0} and it's type is {1} ".format(b,type(b)))
           c = -10
           print("the value of c is {0} and it's type is {1}".format(c,type(c)))
           d = 0
           print("the value of d is {0} and it's type is {1}".format(d,type(d)))
        ```  
                Value of var   | Value after -var
                ---------------------------------
                 Positive      |   Negative
                 Negative      |   Positive
                 Zero          |   Zero  

### Binary Operators
Those operators which works on two operands are called binary operators which is
 operand(_operator_)operand
+ **Addition** $$ a + b$$
  The + operator can be utilized in binary form between the two operands. It
  will produce a sum which can be positive, negative or zero depending on 
  the values of the operands.
  ```python
  a = 1
  b = 2
  print("Adding a and b, we obtain {0}".format(a+b))
  ```
+ **Subtraction** $$a-b$$
  The - operator is utilized to subtract b from a. 
  ```python
  a = 10
  b = 2
  print("Subtracing b from a , we obtain {0}".format(a-b))
  ```
+ **Multiplication**$$a\times b$$ 
  + _a_ and _b_ as **Integer** or **Floating point**:   
   This operand * multiplies _a_ by _b_. 
   ```python 
   a = 1
   b = -10
   print("after multiplying we obtained {0}".format(a*b))
   ```
  + one of the two variables is a **String** and the other is an **Integer**.
  Suppose, a string is named as _a_ and an integer is named as _b_ .Multiplying the string with the integer gives us a string which repeats b times without space.
  ```python
  a = 'fun'
  b = 2
  print("we obtained a string that repeated {0} times which is {1}".format(b,a*b))
  ```
+ **Floor Division** $$ceil(\frac{a}{b})$$

  The operator '//' is used to find floor division between the two operands.   Basically, it gives us the quotient after dividing var1 by var2.
  ```python
  var1 = 20
  var2 = 10
  print("Dividing var1 by var2, we get {0}".format(var1//var2))
  ```
+ **Division**$$\frac{a}{b}$$ 
  The operator '/' gives us the quotient when a is divided by b. Besides, it is of type **float**
+ var1 % var2:   

  This operator %(modulus) gives us the **remainder** after var1 is divided by var2.
  ```python
  print("Remainder is  = {0}".format(var1%var2))
  ```

+ **Exponentiation**  $$a^b$$    
  The operator here means **a** raised to the power **b**.
   Similarly, we can get perform this operation for several variables.
  $$a^{b^{g^{d^{e}}}}$$  
  in **[Python](https://en.wikipedia.org/wiki/Python_(programming_language))**, which can be coded as :  
  ```python
  a = 2 
  b = 3
  g = 1
  d = 2
  e = 1
  print("the value of a^b^g^d^e is {0}".format(a**b**g**d**e))
  ```  

Now, we'll discussion about **Relational Operators**.
Basically, this operator compares values to generate either True or False:
#### These operators are shown by examples in the form operand(_operator_)operand:
+ _Greater than_$$a>b$$
  ```python 
  print("The statement - ""2 is greater than 3"" is {0} and it's type is {1}".format(2>3,type(2>3)))
  ``` 
+ _Less than_ $$a<b$$
```python
 print("The statement - ""2 is smaller than 3"" is {0} and it's type is {1}".format(2<3,type(2<3)))
  ``` 
+ _Equality test_ $$a==b$$
 ```python 
  print("The statement - ""2 equal to 2"" is {0} and it's type is {1}".format(2==2,type(2==2)))
  ``` 
+ _Not equal to_ $$a!=b$$
 ```python 
  print("The statement - ""2 is not equal to 3"" is {0} and it's type is {1}".format(2!=3,type(2!=3)))
 ```
+ _Greater than or equal to_ $$a>=b$$
  ```python 
  print("The statement - ""6 is greater than or equal to 5"" is {0} and it's type is {1}".format(6>=5,type(6>=5)))
  ``` 
+ _Less than or equal to_ $$a<=b$$
  ```python 
  print("The statement - ""2 smaller than or equal to 3"" is {0} and it's type is {1}".format(2<=3,type(2<=3)))
  ``` 

### Now, we'll discuss about **Assignment operators**. Well, it can be in the two forms:
1. _operand_ = _expression_ or _operand_
2. _operand_ (operator)= _expression_ or _operand_

Everyone is familiar with 1. I'll discuss about 2 here.
Is it also possible that, a computation can be done using an operator and assigning that to the operand itself.

#### What I mean is:
+ $$a+=b$$  means a assigning to itself the summation of a and b
+ $$a-=b$$  means a assigning to itself the subtraction of b from a
+ $$a*=b$$  means a assigning to itself the product of a and b.  

Similarly we do that, we other operators

Now, we are going to talk about **Logical Operators**.
#### There are mainly **three** logical operators:
+ and
+ or
+ not
```python 
   print("True and True is {0}".format(True and True))
   print("True and False is {0}".format(True and False))
   print("Not 0 is {0}".format(not 0))
```
We know, logically,
for **_and_** operator,   
 **True** _and_ **True** = **False**  
 **True** _and_ **False** = **False**  
 **False** _and_ **True** = **False**
 **False** _and_ **False** = **False**

------------------------------------------------  

for **_or_** operator,
if either of the **bool** values is **True** , then it is **True**.  

------------------------------------------------

For **_not_** operator,  
we complement the boolean value, so we get the opposite result of the 
input.  For example:  
**_not_** **True** = **False**  
**_not_** **False** = **True**     


-------------------------------------------------

There are multiple operators that can be connected with operands which can be any of the above three logical operators to produce a compound expression.

If there are multiple operands having _or_ operand among themselves, then it's a 
compound _or_ expression.  

x_1 or x_2 or x_3  or ......... x_n = true (if at least one of them is true)

If there are multiple operands having _and_ operand among themselves, then it's a compound _and_ expression.

x_1 and x_2 and x_3 and .......... x_n = true ( if all of them are true)

short-circuit evaluation is something that's done from left to right of an expression.  Evaluation stops as soon as a statement that's true is found.
An expression can be error-prone but ignored even if there is a true statement before that expression in a compound-or-expression. 


## **Bitwise operators**
+ **a & b** gives us the logical _and_ of the bits of the corresponding bits of the operands. 

+ **~a** gives us the logical _not_ of the bits of the corresponding bits of the operand.

+ **a|b** gives us the logical _or_ of the bits of the corresponding bits of the operands.

+ **a^b** gives us the logical _xor_ of the bits of the corresponding bits of the operands.

+ **left shift( a >> b )** gives us _a_ right shifted by _b_ bits.
+ **right shift(a << b)** gives us _a_ left shifted by _b_ bits.
