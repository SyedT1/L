## **Loops**
Loops are used to execute the same block of code over and over again. In a block of code, the statements are executed in order. 
**line 1 -> then line 2 -> and so on.. until the block of code ends**

##### However there are two types of iteration :  
+ **Indefinite**:   

  The condition is **not explicitly** defined in advance.Hence loop ends as soon as a condition is met at some point. for example:
 ```python
 i = 16
 while True:
   if i%15==10:
     print("The loop ends here as {0} is divisible by 15".format(i)
     break
   i = i + 1
 ```
    
+ **Definite**
 The condition is **explicitly** defined in advance. Hence loop ends as soon as the condition meets.
 ```python
 for i in range(1,10):
   print("Number {0}".format(i),end='\n')
 ```     
   
### **While Loop:**
```python
  while <expr>:
    <statement(s)> 
``` 
statement within the while loop is executed until <expr> is False.


```python
  i = 5
  while i<10:
    print(i,end=',')
    i = i + 1
```
As soon as i becomes equal to 10 , the statement within the loop stops getting printed.  

### **Break and Continue**

```python
  for i in range(1,100):
    print("{0} is done".format(i))
    if i%10==0:
      break
    print("hello world")
```
here the **for loop**  breaks because the condition i%10 is met(i==10). Here,the statement below _break_ statement is ignored because the **for loop** also stops executing the block.


```python
  for i in range(1,100):
    print("{0} is done".format(i))
    if i%10==0:
      continue
    print("hello world")
```

Here the **for loop** doesn't execute the statements below _continue_ as soon as i is divisible by 10. All the multiples of 10 is ignored. Hence,
n - floor(range/10) statements of _hello world_ are executed.

### **Else Loop**

using **else** statement after the end of while loop is a wonderful feature in python. Because in other languages there is no feature like that. The reason it is wonderful is because, this feature comes handy in case of some computations.
  
For example:
```cpp
#include <iostream>
using namespace std;
int main(){
  ios_base::sync_with_stdio(false);
  cin.tie(NULL);
  int n;
  while(true){
     cin>>n;
     bool mark = true;
     for(int i = 2;i<n;i++){
         if(n%i==0){
             mark = false;
             break;
         }
     }
     if(mark==true && n>1){
       break;
     }
  }
  cout<<"Finally you got a prime and the value is  = "<<n<<endl;
  return 0;
}
```
in the above c++ code, i had to use a boolean value to mark the number as a composite number.
Now with **Python**, it is quite easy.
```python
  x = 0
while True:
    x = int(input())
    if x>1:
      for i in range(2,x):
        if x%i==0:
          break
      else:
        print("Prime found and it's value is {0}".format(x))
        break
```

here we didn't need a separate boolean value to mark a prime number.

_else_ helps here because when the break statement is executed, the _else_ statement is ignored.

### **One-line while or for loops**

While or for loops has this feature. If there are multiple statements in the block that makes up the loop body, they can be separated by semicolons (;)
```python 
i = 1
for i in range(i,10):print("Number {0}".format(i));i+=1
```
